La ecualizacion de histograma se utiliza para la corrección de imagenes con una mala exposicion. Una correcta exposicion tiene sus intensidades distribuidas de forma uniforme. Al optimizar la distribución de los niveles de brillo se mejora el contraste y por ende se pueden apreciar mas los detalles.  

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena_gray.tif', cv2.IMREAD_GRAYSCALE)

imgmod = cv2.equalizeHist(img) # https://docs.opencv.org/4.x/d4/d1b/tutorial_histogram_equalization.html

# plot
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
plt.title("Imagen")
plt.imshow(img, interpolation='none', cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2)
plt.title("Imagen Modificada")
plt.imshow(imgmod, interpolation='none', cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3)
plt.title("Histograma")
histograma, bins = np.histogram(img.flatten(), bins=256, range=[0, 256])
plt.bar(bins[:-1], histograma, width=1)

plt.subplot(2,2,4)
plt.title("Histograma Modificado")
histograma, bins = np.histogram(imgmod.flatten(), bins=256, range=[0, 256])
plt.bar(bins[:-1], histograma, width=1)

plt.show()
```
