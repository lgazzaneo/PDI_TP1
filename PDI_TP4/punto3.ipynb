BIT PLANE SLICING

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('ej3img.png', cv2.IMREAD_GRAYSCALE)

bitplanes = [(img >> i) & 1 for i in range(8)]

fig, axes = plt.subplots(1, 8, figsize=(20, 5))
for i in range(8):
    axes[i].imshow(bitplanes[i], cmap='gray')
    axes[i].axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('onedollar.jpg', cv2.IMREAD_GRAYSCALE)

bit_planes = [(img >> i) & 1 for i in range(8)]

fig, axes = plt.subplots(1, 8, figsize=(20, 5))
for i in range(8):
    axes[i].imshow(bit_planes[i], cmap='gray')
    axes[i].set_title(f'Bit-plane {i+1}')
    axes[i].axis('off')
plt.show()

# Reconstruir la imagen usando los bit-planes 8, 7, 6, 5 y 1
reconstructed_img = (bit_planes[7] << 7) + (bit_planes[6] << 6) + (bit_planes[5] << 5) + (bit_planes[4] << 4) + (bit_planes[0] << 0)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img, cmap='gray')
axes[0].set_title('Imagen Original')
axes[0].axis('off')
axes[1].imshow(reconstructed_img, cmap='gray')
axes[1].set_title('Imagen Reconstruida (bit-planes 8, 7, 6, 5, 1)')
axes[1].axis('off')

plt.show()
```
