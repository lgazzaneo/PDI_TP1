```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image = np.array(Image.open('./images/eyebw.jpg').convert('L'), dtype=np.float32) / 255.0

c, g = 1.0, 1.5
transformacion1 = c * (image ** g)

transformacion2 = c * (np.log(1+image) )

low, high = 0.3, 0.7 # Contrast Streching como en ej1
transformacion3 = np.piecewise(
    image,
    [image < low, (image >= low) & (image <= high), image > high],
    [0, lambda x: (x - low) / (high - low), 1]
)

titles = ["Original", "Transformacion 1", "Transformacion 2", "Transformacion 3"]
images = [image, transformacion1, transformacion2, transformacion3]

plt.figure(figsize=(12, 5))
for i, (img, title) in enumerate(zip(images, titles), 1):
    plt.subplot(1, 4, i)
    plt.imshow(img, cmap='gray', vmin=0, vmax=1)
    plt.xticks([]), plt.yticks([])
    plt.title(title)
plt.show()
```
