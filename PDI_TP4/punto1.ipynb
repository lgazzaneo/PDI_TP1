```python
import numpy as np
import matplotlib.pyplot as plt


image = np.array(Image.open('./images/lena_gray.tif').convert('L'), dtype=np.float32) / 255.0

height, width = image.shape
negative = 1 - image

transform = np.piecewise(image, 
    [image < 0.25, (image >= 0.25) & (image <= 0.5), image > 0.5],
    [0, lambda x: (x - 0.25) * 4, 1]
)

titles = ["Original", "Negative", "Transform"]
images = [image, negative, transform]

plt.figure(figsize=(12, 5))
for i, (img, title) in enumerate(zip(images, titles), 1):
    plt.subplot(1, 3, i)
    plt.imshow(img, cmap='gray', vmin=0, vmax=1)
    plt.xticks([]), plt.yticks([])
    plt.title(title)

plt.show()
```
