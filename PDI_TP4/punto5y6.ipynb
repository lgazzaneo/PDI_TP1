El brillo es la intensidad promedio de todos los pixeles de la imagen, mientras que el contraste se refiere a la diferencia entre las partes m치s claras y las partes m치s oscuras de la imagen. El brillo ajusta la luminosidad total de una imagen, haciendo que todos los tonos, tanto claros como oscuros, se aclaren o se oscurezcan juntos. El contraste, en cambio, altera la diferencia entre los tonos m치s claros y m치s oscuros de la imagen, aumentando o disminuyendo el rango tonal entre ellos.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('eyebw.jpg', cv2.IMREAD_GRAYSCALE)

image = image.astype(np.float32) / 255.0 # lo paso a float

brillo = 0.5 # ente 0 y 1
contraste = 1

imgmodificada = (image - 0.5) * contraste
imgmodificada = imgmodificada + 0.5 + brillo

imgmodificada = np.clip(imgmodificada, 0, 1)

# plot
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Imagen")
plt.imshow(image, interpolation='none', cmap='gray', vmin=0, vmax=1)
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2)
plt.title("Imagen con brillo")
plt.imshow(imgmodificada, interpolation='none', cmap='gray', vmin=0, vmax=1)
plt.xticks([]), plt.yticks([])

plt.show()

brillo = 0 # ente 0 y 1
contraste = 4

imgmodificada = (image - 0.5) * contraste
imgmodificada = imgmodificada + 0.5 + brillo

imgmodificada = np.clip(imgmodificada, 0, 1)

# plot
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Imagen")
plt.imshow(image, interpolation='none', cmap='gray', vmin=0, vmax=1)
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2)
plt.title("Imagen con contraste")
plt.imshow(imgmodificada, interpolation='none', cmap='gray', vmin=0, vmax=1)
plt.xticks([]), plt.yticks([])

plt.show()
```
