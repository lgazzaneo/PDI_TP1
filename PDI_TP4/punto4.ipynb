Un histograma es un grafico que tiene en el eje horizontal los diferentes niveles de luminosidad (0 - 255 parar una imagen de 8 bits), y en el eje vertical se muestra la cantidad de p√≠xeles que tienen cada intensidad  (su frecuencia absoluta).

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('lena_gray.tif', cv2.IMREAD_GRAYSCALE)

frec_abs = [0]*256 # array que guarda la cantidad de repeticiones de cada valor
width, height = image.shape

# voy pixel por pixel
for i in range(width):
    for j in range(height):
        frec_abs[image[i, j]] += 1

intensidades = np.arange(256)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Imagen")
plt.imshow(image, interpolation='none', cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2)
plt.title("Histograma de la imagen")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia Abs")
plt.bar(intensidades, frec_abs, width=1, align='edge')

plt.show()
```
