A) 
    Un filtro Bayer es una matriz de pequeños filtros de color que se coloca sobre el sensor de las cámaras digitales. Cada filtro deja pasar solo uno de los colores primarios: rojo (R), verde (G) o azul (B). La distribución más común es RGGB, donde hay el doble de filtros verdes que rojos o azules, porque el ojo humano es más sensible al color verde.
    Se utiliza para que el sensor pueda capturar información de color. Cada píxel registra la intensidad de uno de los tres colores, y luego un proceso llamado demosaicing combina esos datos para reconstruir una imagen en color completa.

B)
    El demosaicing es el proceso mediante el cual las cámaras digitales reconstruyen una imagen a color a partir de los datos capturados por un arreglo de filtros de color (CFA). Como cada píxel del sensor registra solo un color, el demosaicing se encarga de completar la información faltante estimando los colores basándose en los píxeles vecinos.
    Existen diversos métodos para realizar este proceso, y la investigación continúa buscando mejoras, especialmente para cámaras de eventos que funcionan de manera distinta a las cámaras tradicionales. Los avances actuales buscan obtener imágenes más nítidas y precisas, incluso cuando parte de la información del sensor es limitada o incompleta, elevando así la calidad final de las fotografías digitales.
