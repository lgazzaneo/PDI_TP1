```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Función para aplicar el filtro Laplaciano y recuperar el fondo
def laplacian_background_recovery(image_path):
    image = cv2.imread(image_path)
    laplacian = cv2.Laplacian(image, cv2.CV_64F)

    laplacian_absoluto = np.absolute(laplacian)
    laplacian_8bit = np.uint8(laplacian_absoluto)

    recovered_background = cv2.subtract(image, laplacian_8bit)

    return image, laplacian_8bit, recovered_background

image_path = 'C:/Users/tomyl/OneDrive/Escritorio/Procesamiento de Imagenes Digitales/venv/TP3/img/moon.png'

# Ejecutar la función y mostrar los resultados
original_image, laplacian_image, background_image = laplacian_background_recovery(image_path)

plt.figure(figsize=(12, 8))

# Mostrar las imágenes

plt.subplot(1, 3, 1)
plt.title('Imagen Original')
plt.imshow(original_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Laplaciano')
plt.imshow(laplacian_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Mas fondo')
plt.imshow(background_image, cmap='gray')
plt.axis('off')

plt.show()
```

El procedimiento aplicado a la imagen de la luna se compone de tres fases principales:

1. Carga de la imagen original.
2. Aplicación del operador Laplaciano.
3. Resta del resultado del Laplaciano a la imagen original para generar una versión con mayor presencia de fondo.

Como resultado, se obtienen tres imágenes distintas: la primera es la original, que muestra la superficie lunar y sus cráteres sin alteraciones; la segunda corresponde a la transformada mediante el operador Laplaciano, que enfatiza los bordes y los detalles finos mostrando zonas brillantes donde hay variaciones abruptas de intensidad y oscuras en áreas uniformes; y la tercera es la imagen con el fondo realzado, obtenida al sustraer el Laplaciano de la original, lo que produce un efecto de suavizado en los bordes, disminuye el contraste y resalta las zonas más uniformes de la superficie, reduciendo además el ruido de alta frecuencia.

Este método ofrece diversas formas de observar el mismo objeto astronómico, permitiendo ajustar el nivel de detalle y contraste según las necesidades del análisis o de la representación visual.
