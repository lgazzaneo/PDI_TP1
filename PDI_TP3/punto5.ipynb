```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

original_image = cv2.imread('C:/Users/tomyl/OneDrive/Escritorio/Procesamiento de Imagenes Digitales/venv/TP3/img/blurry_moon.tif', cv2.IMREAD_GRAYSCALE)
blurred_image = cv2.imread('C:/Users/tomyl/OneDrive/Escritorio/Procesamiento de Imagenes Digitales/venv/TP3/img/blurry_moon.tif', cv2.IMREAD_GRAYSCALE)

A = 1.3

blurred_image = cv2.resize(blurred_image, (original_image.shape[1], original_image.shape[0]))

laplacian_blurred = cv2.Laplacian(blurred_image, cv2.CV_64F)

hi_boost_image = A*original_image - laplacian_blurred

hi_boost_image = np.clip(hi_boost_image, 0, 255).astype(np.uint8)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.subplot(1, 2, 2)
plt.imshow(hi_boost_image, cmap='gray', vmin=0, vmax=255)
plt.title(f'Hi-Boost with Laplacian of Pre-Blurred (A={A})')
plt.xticks([]), plt.yticks([])

plt.show()
```

Este ejercicio aplica el filtro Hi-Boost para mejorar una imagen borrosa de la luna.

Proceso:

1. Se carga la imagen borrosa.
2. Se calcula su Laplaciano.
3. Se aplica la frmula del filtro Hi-Boost: IHB=AIo2Ib (A=1.3)

Se ajustan los valores de los pxeles al rango [0, 255].

Resultados:

1. La imagen original presenta poca definicin.
2. La imagen filtrada muestra una mayor nitidez y contraste, con bordes y detalles ms marcados, aunque con posible aumento de ruido.

Conclusiones:
El filtro Hi-Boost mejora notablemente la claridad y el contraste de la imagen lunar. El valor A=1.3 ofrece un equilibrio adecuado entre realce y ruido. En general, esta tcnica resulta til para reforzar detalles en imgenes astronmicas obtenidas con baja calidad o resolucin limitada.
