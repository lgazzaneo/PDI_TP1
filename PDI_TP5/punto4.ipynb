```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.signal import convolve2d

def mse(original, restored):
    return np.mean((original - restored) ** 2)

image = cv2.imread('clown.jpg',cv2.IMREAD_GRAYSCALE)

n = 8
h = np.eye(n) / n

H = np.fft.fft2(h, s=image.shape)
H_shifted = np.fft.fftshift(H)

blur = convolve2d(image, h, mode = 'same')

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(image, interpolation='none', cmap = 'gray')
plt.title('Original')
plt.subplot(122)
plt.imshow(blur, interpolation='none', cmap = 'gray')
plt.title('Motion Blur')
plt.show()

mseimg = mse(image,blur)
print("=== Mean Squared Error (MSE) ===")
print(f"  MSE:  {mseimg:.6f}")
```
